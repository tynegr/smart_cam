```python
import pandas as pd
import numpy as np

# Функция для поиска самого правого ненулевого значения
def find_rightmost_nonnull(row):
    columns = ['Param4', 'Param3', 'Param2', 'Param1', 'subcategory_name', 'category_name']
    for col in columns:
        if pd.notna(row[col]):
            return row[col]
    return np.nan

# Применение функции к DataFrame
df['category'] = df.apply(find_rightmost_nonnull, axis=1)

df
```

```python
final_df = df[['external_image_id', 'category', 'price']]
final_df
```

```python
final_df.to_csv('final_df.csv')
```
